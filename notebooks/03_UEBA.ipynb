## Planned scope

- Behavior/entity profiling from flow-level features already produced in notebook 01 (per-source-IP aggregates: connection counts, byte volumes, service diversity over rolling windows).
- Anomaly scoring via the unsupervised detectors trained in notebook 02 (IsolationForest/OCSVM), re-applied at the entity level rather than the flow level.
- Sequence analysis: order flows per entity into time-ordered sequences; candidate approach is n-gram / Markov-chain transition-probability scoring (no external dependency), with an LSTM-based sequence scorer as a GPU-requiring upgrade (see deep_stubs.FlowLSTM).
- Graph-based behavior analytics: build an entity-interaction graph (networkx) from source/destination pairs; flag high-centrality or newly-appearing nodes.
