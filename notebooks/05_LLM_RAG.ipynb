## Planned scope

- Prompt templates for incident summarization, MITRE technique explanation, and confidence estimation.
- RAG pipeline: embed the knowledge graph's textual descriptions (notebook 04) with sentence-transformers, index with FAISS, retrieve top-k context per incident.
- Hallucination mitigation: constrain generation to retrieved context; require the model to cite which retrieved passage supports each claim; flag ungrounded claims for analyst review.
