## Planned scope

- Network topology simulator: networkx graph of hosts/subnets/services parameterized from the entity graph in notebook 03.
- Attack simulator: rule-based propagation model (e.g. SIR-style spread) seeded with ATT&CK techniques from notebook 04.
- Defense simulator: patch/isolate/monitor actions as graph mutations; risk estimated as expected-compromised-hosts under a given defense policy.
- What-if / scenario replay: re-run the propagation model under alternate initial conditions or defense policies for comparison.

## Feasibility in this environment

The topology + rule-based propagation model is executable today with networkx/numpy only. A full discrete-event simulation (simpy) is a natural upgrade once the rule-based version is validated.
