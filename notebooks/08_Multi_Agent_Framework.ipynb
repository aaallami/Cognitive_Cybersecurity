## Planned scope

- Agent roles: Triage Agent (notebook 03 output -> priority queue), KG Agent (notebook 04 lookups), RAG/Explainer Agent (notebook 05), Twin/Policy Agent (notebooks 06-07), Human-Interface Agent (notebook 09).
- Orchestration via LangGraph (explicit state graph) as the primary target, with a CrewAI variant documented as an alternative.
- Task decomposition: an incoming alert is decomposed into sub-tasks routed to the relevant agent(s), with results aggregated back into a single incident report.

## Feasibility in this environment

Requires an LLM backend (05) to be live, since the agents' reasoning steps are LLM calls; the orchestration graph structure itself (LangGraph StateGraph) can be defined and unit-tested with mocked LLM responses even without live credentials.
