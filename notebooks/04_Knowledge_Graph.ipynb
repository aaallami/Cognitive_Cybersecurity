## Planned scope

- MITRE ATT&CK technique/tactic nodes, loaded from a locally cached STIX bundle (download once from https://github.com/mitre/cti, cache under datasets/threat_intel/).
- CVE/CWE/CAPEC cross-references via the NVD API (requires network access to services.nvd.nist.gov, not in this sandbox's allowlist) or a locally cached NVD JSON feed.
- Attack graph construction (networkx) linking observed anomaly categories (from notebook 03) to plausible ATT&CK techniques via a rule-based mapping table.
- Reasoning: shortest-path / centrality analysis over the attack graph to rank likely attack progressions.

## Feasibility in this environment

Fully offline-capable once the MITRE ATT&CK STIX bundle is cached locally; the NVD cross-reference step needs either network access this sandbox lacks, or a pre-downloaded NVD feed supplied by the user.
